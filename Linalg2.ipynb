```python
import numpy as np
from scipy.linalg import solve
from math import exp, sin
import matplotlib.pyplot as plt
```

```python
def f(x):
    return sin(x / 5) * exp(x / 10) + 5 * exp(-x / 2)
x = np.linspace(1, 15, 45)
y = np.array([f(i) for i in x])
```

Многочлен первой степени

```python
xcheck_1 = [1., 15.]
A1 = np.mat([[i**step for step in range(2)] for i in xcheck_1])
b1 = np.array([[f(1.)], [f(15.)]])
print A1, b1
```

```python
w = solve(A1, b1)
print w
np.dot(A1, w) == b1
```

```python
ycheck_1 = np.array([w[0, 0] + w[1, 0] * i for i in xcheck_1])
print ycheck_1
```

```python
plt.plot(x, y, label='func')
plt.plot(xcheck_1, ycheck_1, label='check')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title("First")
plt.legend()
```

Многочлен второй степени

```python
xcheck_2 = [1., 8., 15.]
A2 = np.array([[i**step for step in range(3)] for i in xcheck_2])
b2 = np.array([[f(1.)], [f(8.)], [f(15.)]])
print A2, b2
```

```python
w_2 = solve(A2, b2)
print w_2
```

```python
ycheck_2 = np.array([w_2[0, 0] + w_2[1, 0] * i + w_2[2, 0] * i**2 for i in xcheck_2])
print ycheck_2
```

```python
plt.plot(x, y, label='func')
plt.plot(xcheck_2, ycheck_2, label='check')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title("Second")
plt.legend()
```

Многочлен третьей степени

```python
xcheck_3 = [1., 4., 10., 15.]
A3 = np.array([[i**step for step in range(4)] for i in xcheck_3])
b3 = np.array([[f(1.)], [f(4.)], [f(10.)], [f(15.)]])
print A3, b3
```

```python
w_3 = solve(A3, b3)
print w_3
```

```python
ycheck_3 = np.array([w_3[0, 0] + w_3[1, 0] * i + w_3[2, 0] * i**2 + w_3[3, 0] * i**3 for i in xcheck_3])
print ycheck_3
```

```python
plt.plot(x, y, label='func')
plt.plot(xcheck_3, ycheck_3, label='check')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title("Third")
plt.legend()
```
