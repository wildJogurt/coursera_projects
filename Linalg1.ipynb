```python
from scipy import linalg
from scipy.spatial import distance
import numpy as np
import re
from collections import Counter
```

```python
with open('sentences.txt', 'r') as f:
    read_data = f.read()
cat_sentences = read_data.lower()
cat_cut = cat_sentences.split('\n')[:-1]
cat_cut = [re.findall('[a-z]+-[a-z]+|[a-z]+', i) for i in cat_cut]
cat_sentences = re.sub(r"\n", " ", cat_sentences)
cat_sentences
```

```python
cat_words = re.findall('[a-z]+-[a-z]+|[a-z]+', cat_sentences)
cat_words
```

```python
cats_counter = Counter(cat_words)
cats_words = map(list, cats_counter.items())
cats_words
```

```python
cat_matrix = np.zeros((22, 254))
for sentence in range(len(cat_matrix)):
    for word in range(len(cat_matrix[0])):
        for i in cat_cut[sentence]:
            if cats_words[word][0] == i:
                cat_matrix[sentence][word] += 1
```

```python
for i in cat_matrix:
    print i
```

```python
for j in range(22):
    for i in range(254):
        if cat_matrix[j][i] > 0:
            print cats_counter.items()[i], cat_matrix[j][i]
    print '--------------------------------------\n--------------------------------------\n--------------------------------------'
```

```python
cos = []
for sentence in cat_matrix:
    cos.append(distance.cosine(cat_matrix[0], sentence))
cos
```

```python
min_cos1 = cos[1]
min_cos2 = cos[1]
for i in cos[1:]:
    if i < min_cos1:
        min_cos2 = min_cos1
        min_cos1 = i
    elif min_cos2 < i < min_cos1:
        min_cos2 = i
min_cos1, min_cos2
```
